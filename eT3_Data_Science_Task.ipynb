# Q_1_Which drink has the highest calories from the dataset?


```python
df['B_category'].unique()
```

```python
df['Beverage'].unique()
```

```python
DF = pd.DataFrame()
```

```python
DF['Category']= df['B_category']
DF['Calories'] = df['Calories']
```

```python
DF
```

```python
cal = DF.sort_values(by='Calories', ascending= False)
cal
```

```python
fig = px.bar(cal, x='Category', y='Calories' , color='Calories')
fig.show()
```

So the highest calories from Beverage_category is 510 from "Signature Espresso Drinks"

```python
df[df['B_category']=="Signature Espresso Drinks"].shape
```

```python
DF_2 = pd.DataFrame()
```

```python
DF_2 = df[df['B_category']=="Signature Espresso Drinks"]
```

```python
DF_2
```

```python
DF_3 = pd.DataFrame()
```

```python
DF_3['Beverage']= DF_2['Beverage']
DF_3['Calories'] = DF_2['Calories']
```

```python
DF_3
```

```python
cal_2 = DF_3.sort_values(by='Calories', ascending= False)
cal_2
```

```python
fig = px.bar(cal_2, x='Beverage', y='Calories' , color='Calories')
fig.show()
```

```python
# data visualization to see which signature espresso drink has the highest calories
sns.barplot(x="Beverage", y="Calories", data=DF_3)
plt.xticks(rotation=30, ha='right')
plt.title("Signature Espresso Drinks")
```

So the drink has the highest calories is "White Chocolate Mocha (Without Whipped Cream)"

```python
df.sort_values('Calories', ascending=False)
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x="Beverage", y="Calories", data=df)
plt.xticks(rotation=40, ha='right')
plt.title("Beverage Calories")
plt.show()
```

So the drink has the highest calories is "White Chocolate Mocha (Without Whipped Cream)"
